```python
import pydicom
import os
import matplotlib.pyplot as plt
from collections import Counter
```

```python
pet_positions = []
ct_positions = []
directory = "/home/filip/Documents/PREŠERNOVA NALOGA/kontrole/5KBJJ_20230415_162007/"
for file in os.listdir(directory):
    if file.endswith(".IMA"):
        slice = pydicom.dcmread(f"{directory}/{file}")

        if all([x in slice.SeriesDescription for x in ["OSEM", "4mm"]]):
            if "ImagePositionPatient" in dir(slice):
                if float(slice.ImagePositionPatient[2]) == -416.0:
                    plt.imshow(slice.pixel_array, cmap=plt.cm.bone)
                    plt.show()                
                pet_positions.append(slice.ImagePositionPatient[2]) 

        if all([x in slice.SeriesDescription for x in ["CT", "rostat"]]):
            if "ImagePositionPatient" in dir(slice):
                ct_positions.append(slice.ImagePositionPatient[2])     
```

```python
total = 0
for pos in pet_positions:
    if pos in ct_positions:
        total += 1
```
